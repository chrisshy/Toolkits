```python
url = "https://music.163.com/#/playlist?id=5007845732"
chromeOptions = webdriver.ChromeOptions()
chromeOptions.add_argument('--headless')
chromeOptions.add_argument('--no-sandbox')
chromeOptions.add_experimental_option('useAutomationExtension', False)
capabilities = {'chromeOptions': {'useAutomationExtension': False, 'args': ['--disable-extensions']}}
driver = webdriver.Chrome(executable_path='C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe',options=chromeOptions)#,desired_capabilities=capabilities)

wb = Workbook()    #创建文件对象
ws = wb.active     #获取第一个sheet

for i in range(1304):
    driver.get(url_list[i])
    #driver.find_element_by_xpath('//*[@id="treeDemo_2_span"]').click()
    #driver.switch_to.default_content()
    #frame = driver.find_elements_by_tag_name('iframe')[1]
    driver.switch_to.frame("contentFrame")
    #driver.find_element_by_id("pageSea").send_keys("544")
    #driver.find_element_by_xpath('//*[@id="fwd_2"]/a').click()

    time.sleep(1.5)
    
    
    content=[]

    #WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="auto-id-5AkinjNjpOuAO8u"]/div[2]/div/div[1]/div/h2')))
    # 标题
    list_contents=driver.find_elements_by_css_selector("div.cnt > div > div.hd.f-cb > div > h2")
    for contents in list_contents:
        title = contents.text
        content.append(title)

    # 作者
    list_contents=driver.find_elements_by_css_selector("div.cnt > div > div.user.f-cb > span.name > a")
    for contents in list_contents:
        title = contents.text
        content.append(title)

    # 创建时间
    list_contents=driver.find_elements_by_css_selector("div.cnt > div > div.user.f-cb > span.time.s-fc4")
    for contents in list_contents:
        temp = contents.text
        content.append(re.findall(r'^\d{4}-\d{1,2}-\d{1,2}',temp)[0])    

    #收藏量
    list_contents=driver.find_elements_by_xpath('//*[@id="content-operation"]/a[3]/i')
    for contents in list_contents:
        temp = contents.text
        try:
            content.append(re.findall(r'\d+',temp)[0]) 
        except:
            content.append(0)

    #转发量
    list_contents=driver.find_elements_by_css_selector('#content-operation > a.u-btni.u-btni-share > i')
    for contents in list_contents:
        temp = contents.text
        try:
            content.append(re.findall(r'\d+',temp)[0]) 
        except:
            content.append(0)

    # 下载量
    list_contents=driver.find_elements_by_css_selector('#content-operation > a.u-btni.u-btni-dl > i')
    for contents in list_contents:
        temp = contents.text
        try:
            content.append(re.findall(r'\d+',temp)[0]) 
        except:
            content.append(0)

    # 评论量
    list_contents=driver.find_elements_by_css_selector('#cnt_comment_count')
    for contents in list_contents:
        temp = contents.text
        try:
            content.append(re.findall(r'\d+',temp)[0]) 
        except:
            content.append(0)

    #标签
    list_contents=driver.find_elements_by_css_selector('div.cnt > div > div.tags.f-cb > a > i')
    text=''
    for contents in list_contents:
        temp = contents.text
        text=text+" "+temp
    content.append(text) 

    #介绍
    list_contents=driver.find_elements_by_css_selector('#album-desc-more')
    for contents in list_contents:
        temp = contents.text
        content.append(temp) 

    #有多少歌
    list_contents=driver.find_elements_by_css_selector('#playlist-track-count')
    for contents in list_contents:
        temp = contents.text
        content.append(temp)

    # 播放量
    list_contents=driver.find_elements_by_css_selector('#play-count')
    for contents in list_contents:
        temp = contents.text
        content.append(temp)

    # 精彩评论
    # 选取精彩评论10条，不足则由最新评论替换，再不足则取空值
    list_contents=driver.find_elements_by_css_selector('div.cntwrap > div:nth-child(1) > div')
    try:
        content.append(list_contents[0].text)
    except:
        content.append("")
    try:
        content.append(list_contents[1].text)
    except:
        content.append("")
    try:
        content.append(list_contents[2].text)
    except:
        content.append("")
    try:
        content.append(list_contents[3].text)
    except:
        content.append("")
    try:
        content.append(list_contents[4].text)
    except:
        content.append("")
    try:
        content.append(list_contents[5].text)
    except:
        content.append("")
    try:
        content.append(list_contents[6].text)
    except:
        content.append("")
    try:
        content.append(list_contents[7].text)
    except:
        content.append("")
    try:
        content.append(list_contents[8].text)
    except:
        content.append("")
    try:
        content.append(list_contents[9].text)
    except:
        content.append("")

    print(content)
    ws.append(content)
    print(i)
    print("**************YEAH**************")
    if i %100 ==0:
        wb.save("./NetEase cloud music-1.xlsx")
        time.sleep(10)
wb.save("./NetEase cloud music-1.xlsx")
```
