# Realized Volatility & Implied Volatility

值得注意的是波动率通过不同的方法计算得出的意义是不同的，历史波动率反映的是标的资产在过去一段时间的已实现的波动，而在B-S期权定价模型中反求出的波动率是隐含波动率，两者的意义区别在于后者因为看涨期权的存在，体现的是对于未来的前瞻型波动率展望。对于交易员而言，通常也是采用历史时间序列模型反求股价隐含波动率，而本文研究的期望为预测，并期望是获得更为贴近实际预期的价格，因此本文的波动率可以用隐含波动率来替代研究，在拟合波动率时，将隐含波动率作为真实波动率是合理的。

## Realized Volatility

```python
# 计算对数收益率的实现波动
def realized_volatility(df):
    df_n = df.copy()
    df_n['cumsum'] = df_n['Log_Return'].cumsum()
    df_n['total'] = df_n['Log_Return'].sum()
    df_n['total-cumsum'] = df_n['total']-df_n['cumsum']
    df_n['Realized_Volatility']=0
    length = len(df_n)
    for i in range(0,length):
        number = length-1-i
        average = df_n.iloc[i,3]/number
        sum_diff = 0
        for j in range(i+1,length):
            sum_diff = sum_diff + (df_n.iloc[j,0]-average)**2
        if number > 0:
            rv = np.sqrt(sum_diff/number)
        else:
            rv = 0
        df_n.iloc[i,4]=rv
    return df_n
```

## Implied Volatility

根据B-S公式，不难发现看涨期权的价值是股票价格、执行价格、无风险利率、距离到期日时长和隐含波动率的函数。在计算隐含波动率，一般通过该函数逆推。但是在逆推计算时一般是不能直接获得解析解的，所以在这里本文采用牛顿迭代法将波动率作为未知数求得方程数值解。牛顿迭代法的迭代关系式为：

$X_{(n+1)}=X_n-(f(X_n))/(f^1(X_n))   $


```python
# Implied Volatility 
from scipy.stats import norm
import numpy as np
def bscall(S,K,r,sigma,t):
    d1=(np.log(S/K)+(r+0.5*sigma**2)*t)/(sigma*np.sqrt(t))
    d2=d1-sigma*np.sqrt(t)
    return S*norm.cdf(d1)-K*np.exp(-r*t)*norm.cdf(d2)
def bsput(S,K,r,sigma,t):
    d1=(np.log(S/K)+(r+0.5*sigma**2)*t)/(sigma*np.sqrt(t))
    d2=d1-sigma*np.sqrt(t)
    return -S*norm.cdf(-d1)+K*np.exp(-r*t)*norm.cdf(-d2)

def newton_call(P,S,K,r,t):
    sigma=0.5
    #n = 0
    while abs(bscall(S,K,r,sigma,t)-P)>0.1:
        if bscall(S,K,r,sigma,t)>P:
            sigma-=0.0001
        else:
            sigma+=0.0001
        #n = n + 1
        #print(n,sigma)
    return sigma
def newton_put(P,S,K,r,t):
    sigma=0.3
    while abs(bsput(S,K,r,sigma,t)-P)>0.1:
        if bsput(S,K,r,sigma,t)>P:
            sigma-=0.01
        else:
            sigma+=0.01
    return sigma
#print('看涨期权隐含波动率为{:.3f}'.format(newton_call(9,100,100.4,0.04,0.3)))
```

## ARIMA-ACF/PACF/acorr_ljungbox-ARCH

在对股票价格建立均值模型前，必须对其自相关性和偏自相关性进行考察，这是为了确认建立均值模型时滞后阶数的选择，通常而言，常见的均值模型包括ARIMA模型和常数均值模型。在二者选择时需要判断ARIMA模型的p，q阶数，在这里我们通过自相关图和偏自相关图进行初步判断

从图中我们不难发现图形存在双截尾的情况，因此可以认为该对数序列不存在明显的自相关性现象，为了使得判断能够同样从数学计算的角度科学合理化，本文进一步采用Ljung-Box检验进行假设检验，该检验方法即LM检验，通常用来检验各阶数滞后范围内序列的自相关性或序列是否为白噪声，该检验方法的原假设是序列是白噪声序列

从检验结果中可以看到各阶数的p值都远大于5%，因此接受原假设，序列为白噪声序列，不存在相关性，因此此处均值模型的建立采用常数均值模型

从检验结果可以看出各阶数都显著得小于5%，这意味着残差序列的平方是非白噪声序列，存在相关性关系，即存在ARCH效应，应当建立ARCH模型

# GJR-GARCH

许多实证研究证明收益率残差的正负对波动率存在非对称性效应，这种影响又称作杠杆效应。最早由Glosten、Jagannathan与Runkel（1989）提出了GJR模型，他们在GARCH中加入负冲击的杠杆效应，GJR-GARCH模型建模如下

$σ_t^2=α_0+α_1 ε_{(t-1)}^2+γε_{(t-1)}^2 I_[ε_{(t-1)}<0] +β_1 α_{(t-1)}^2 $

式中γ为负冲击的影响系数，$I_{[ε_(t-1)<0]}$ 为分类函数，当$[ε_(t-1)<0]$时，该函数值为1，反之为0。


```python
from arch import arch_model
#am = arch_model(result_002460_train,mean='CONSTANT',vol='GARCH')
am = arch_model(result_600998_train,mean='CONSTANT',p=1, o=1, q=1)
res = am.fit(update_freq=5,options={'maxiter': 500})
print(res.summary())
```

# EWMA

EWMA全称为指数加权移动平均模型，其给予了数值随着距离目标数据点的延长而呈指数型下降的加权系数，这意味着短期的信息权重更大，长期的信息会被逐渐遗忘，这也就减少了过去的波动率信息对现在波动率的影响。EWMA高效而简介，相较而言运算和迭代的速度快，EWMA的建模如下：

$σ_t^2=δσ_{(t-1)}^2+(1-δ) ε_{(t-1)}^2$  $ (0<δ<1)$
