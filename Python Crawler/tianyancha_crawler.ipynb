```python
chromeOptions = webdriver.ChromeOptions()
# chromeOptions.add_argument('--headless')
chromeOptions.add_argument('--no-sandbox')
chromeOptions.add_experimental_option('useAutomationExtension', False)
capabilities = {'chromeOptions': {'useAutomationExtension': False, 'args': ['--disable-extensions']}}
driver = webdriver.Chrome(executable_path='C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe',options=chromeOptions,desired_capabilities=capabilities)

# # 登陆

# 爬取变更信息的excel
wb = Workbook()    
ws = wb.active
# 爬取股东信息的excel
wb2 = Workbook()    
ws2 = wb2.active

# 记录没有爬取成功的url
url_try_again_list = [] 
company_try_again_list = []
a = 1

for url,company in zip(url_list,company_list):
    try:
        driver.get(url)

    #     driver.add_cookie(cookies_1)
    #     driver.add_cookie(cookies_2)
        
        if a == 1:
            time.sleep(40)

        try:
            WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="nav-main-holderCount"]/span[1]')))

            # 爬取该企业的股东信息
            holder_list = driver.find_elements_by_xpath('//*[@id="_container_holder"]/table/tbody/tr')
            holder_number = len(holder_list)
            for i in range(holder_number):
                # 序号
                index = driver.find_element_by_xpath('//*[@id="_container_holder"]/table/tbody/tr['+str(i+1)+']/td[1]').text
                # 股东公司名称
                name = driver.find_element_by_xpath('//*[@id="_container_holder"]/table/tbody/tr['+str(i+1)+']/td[2]/table/tbody/tr/td[2]/a').text
                # 股东类型
                holder_type = driver.find_element_by_xpath('//*[@id="_container_holder"]/table/tbody/tr['+str(i+1)+']/td[2]/table/tbody/tr/td[2]/div').text
                # 持股比例
                rate = driver.find_element_by_xpath('//*[@id="_container_holder"]/table/tbody/tr['+str(i+1)+']/td[3]/div/div/span').text
                # 认缴出资额
                amount = driver.find_element_by_xpath('//*[@id="_container_holder"]/table/tbody/tr['+str(i+1)+']/td[4]/div/span').text
                # 认缴出资日期
                datestr = driver.page_source
                date = re.findall('<!--<span>([0-9-]*)</span>-->',datestr)[0]
                holder_info = [company,index,name,holder_type,rate,amount,date]
                ws2.append(holder_info)
        except:
            holder_info = [company,'','','','','','']
            ws2.append(holder_info)

        # 爬取变更信息
        # 该企业爬取变更信息时需要翻多少页或者不翻页    
#         total_info_num = driver.find_element_by_xpath('//*[@id="_input_changeinfo_total"]').get_attribute('value')
#         count = math.floor(int(total_info_num)/10)
#         if count == 0:
#             # 获取变更信息的content
#             # 一页有多少行信息
#             content_list = driver.find_elements_by_xpath("//*[@id='_container_changeinfo']/div/table/tbody/tr")
#             # num = len(content_list)
#             # for i in range(num):
#             for single_information in content_list:
#                 content_info = [company]
#                 info_all = single_information.find_elements_by_tag_name("td")
#                 for info in info_all:
#                     content_info.append(info.text)
#                 ws.append(content_info)
#         else:
        for i in range(1000):
            content_list = driver.find_elements_by_xpath("//*[@id='_container_changeinfo']/div/table/tbody/tr")
            for single_information in content_list:
                content_info = [company]
                info_all = single_information.find_elements_by_tag_name("td")
                for info in info_all:
                    content_info.append(info.text)
                ws.append(content_info)
            try:
                driver.find_element_by_xpath('//*[@id="_container_changeinfo"]').find_element_by_css_selector('a[class="num -next"]').click()
                time.sleep(3+4*random.random())
            except:
                break

    except:
        url_try_again_list.append(url)
        company_try_again_list.append(company)
        wb.save("./changeinfo_total1031"+str(a)+".xlsx")
        wb2.save("./holderinfo_total1031"+str(a)+".xlsx")
        print(str(a)+'爬取失败！')
    
    time.sleep(10+10*random.random())
    print(a)
    a += 1

wb.save("./126/changeinfo_total1031.xlsx")
wb2.save("./126/holderinfo_total1031.xlsx")
```
